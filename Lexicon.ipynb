```python
import pandas as pd

data = pd.read_csv("hasil_preprocessing.csv")

data['Stemming'] = data['Stemming'].apply(lambda x: " ".join(eval(x)) if isinstance(x, str) and x.startswith('[') else x)

positive_lexicon = set(pd.read_csv("positive.tsv", sep="\t", header=None)[0].str.strip())
negative_lexicon = set(pd.read_csv("negative.tsv", sep="\t", header=None)[0].str.strip())

def determine_sentiment(text):
    words = text.split()
    positive_count = sum(1 for word in words if word in positive_lexicon)
    negative_count = sum(1 for word in words if word in negative_lexicon)
    if positive_count > negative_count:
        return "Positif"
    else:
        return "Negatif"

data['Sentiment'] = data['Stemming'].apply(determine_sentiment)

data['Stemming'] = data['Stemming'].apply(lambda x: x.split() if isinstance(x, str) else x)

data.head(10)
```

```python
Sentiment_counts = data.Sentiment.value_counts()
Sentiment_counts
```

```python
data.to_csv('hasil_lablex.csv', index=False)
```
